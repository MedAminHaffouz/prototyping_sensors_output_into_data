## Importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load for me this data

```python
df=pd.read_csv("./sensor_log.csv")
print(df.describe())
print(df)
```

## Plot it

```python
plt.plot(df['timestamp'], df['potentiometer'],color='green', linestyle='--', marker='o', linewidth=2)
plt.xticks([])  # This hides x-axis labels
plt.title("Potentiometer values")
plt.show()
```

```python
plt.plot(df['timestamp'], df['temperature'],color='red', linestyle='--', marker='o', linewidth=2)
plt.xticks([])  # This hides x-axis labels
plt.title("Temperature values")
plt.show()
```

```python
plt.plot(df['timestamp'], df['humidity'],color='blue', linestyle='--', marker='o', linewidth=2)
plt.xticks([])  # This hides x-axis labels
plt.title("Humidity values")
plt.show()
```

```python
import seaborn as sns

# Select the desired columns
selected_cols = df[['potentiometer', 'temperature', 'humidity']]

# Compute correlation matrix
corr = selected_cols.corr()

# Plot heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
